# Model Analysis

The goal of this notebook is to study how open source packages can be used to perform analysis of engineering models.  Specifically, we will use a simple six dimensional model that produces an estimate of building damage due to high speed winds.  This model is not physically realistic, but it will provide a good framework for testing out the methods available in python.

First we will perform local and global sensitivity analysis.

Second we assume that our model takes hours to run.  This is not the case for this demo, but is likely the case for any accurate physics based model of a true engineering component.  To perform any type of statistical analysis of our system, we must evaluate this model many times, and so we will explore ways to construct metamodels of our system to reduce computational cost.

Third, we will calibrate our model using data and some Bayesian inference techniques for quantifying the uncertainty of our model parameters and inputs.

Fourth, we will use this model to propagate uncertainty through our system using available surrogate models and compare it to the "true" model evaluations.

Lets import the modules that we will use to analyze our system.

```python
import numpy as np

import tensorflow as tf

import chaospy as cp 

import tensorflow_probability as tfp

import scipy
import scipy.stats

tfd = tfp.distributions

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

sns.set(color_codes = True)

import SALib
from SALib.sample import saltelli
from SALib.plotting.bar import plot as barplot

from SALib.sample import fast_sampler
from SALib.analyze import fast
```

```python
# And let's define our system
def FEM_Model_gable_roof(x):
    ## FEM_MODEL_gable_roof
    # creates a scalar valued damage estimate with range [0, Inf]
    # from a four dimensional input space
    ## x[:,0] = building aspect ratio (height / width): unitless.  Range [0, 2]
    ## x[:,1] = wind speed : miles per hour.  Range [0, 300]
    ## x[:,2] = wind angle : radians.  Range [0, pi/2]
    ## x[:,3] = design_speed : mile per hour. Range [0, 300]
    adjustment_factor = 1
    return np.cos(x[:,2])*x[:,1]*x[:,1]*x[:,0]/(adjustment_factor*x[:,3]*x[:,3])
```

```python
# Lets take a look at the model output for a specific building (fixed aspect ratio, fixed design speed).
fixed_design_speed = 160 # Miles per hour

fixed_aspect_ratio = 1 # unitless

# And lets study the damage as a function of wind speed for a fixed wind angle
n_wind_speeds = 20

wind_speed_set = np.linspace(0, 300, num = n_wind_speeds)

fixed_wind_angle = np.pi/16

# lets package the parameters together
parameter_set   = np.zeros([n_wind_speeds, 4])
parameter_set[:, 0] = fixed_aspect_ratio
parameter_set[:, 1] = wind_speed_set
parameter_set[:, 2] = fixed_wind_angle
parameter_set[:, 3] = fixed_design_speed

# compute the output
damage_ratio_set = FEM_Model_gable_roof(parameter_set)

fig = plt.figure(figsize = [16, 8], dpi = 600)
plt.plot(wind_speed_set, damage_ratio_set)
plt.xlabel("Wind Speed")
plt.ylabel("Estimated Damage")
```

## Sensitivity Analysis
### Sobol Sensitivity Indices

```python
## Lets use SALib to do an initial sensitivity analysis
from SALib.sample import saltelli
# define the domain of the problem space.
dims_of_parameter_space = 4
problem = {'num_vars': dims_of_parameter_space,
          'names' : ["Aspect Ratio", "Wind Speed", "Wind Angle", "Design Speed"],
          'bounds': [[0, 2],
                    [0, 300],      # Note: These bounds are defined in FEM_Model_gable_roof(...)
                    [0, np.pi/2],
                    [0, 300]]
          }

# generate a space filling design of experiments
# Saltelli sampler generates N*(2*D + 2) samples, where N is the saltelli param number, and
# D in the dimensionality of the parameter space.
saltelli_parameter_number = 1000               # N(2*4 + 2) = N*(10) = 1000 samples
param_values = saltelli.sample(problem, saltelli_parameter_number)
param_values.shape                            # Verify our result is (1000 x 4) numpy array
```

```python
damage_set = FEM_Model_gable_roof(param_values)
```

```python
from SALib.analyze import sobol
Si = sobol.analyze(problem, damage_set)
print(Si.keys())
```

```python
saltelli_param_set = range(100, 20000, 50)

start_num = len(saltelli_param_set)

first_order_indices = np.zeros([start_num, 4])
second_order_indices = np.zeros([start_num, 4])
total_effect_indices = np.zeros([start_num, 4])


for idx, param in enumerate(saltelli_param_set):
    saltelli_parameter_number = param           
    param_values = saltelli.sample(problem, saltelli_parameter_number)
    damage_set = FEM_Model_gable_roof(param_values)  
    Si = sobol.analyze(problem, damage_set, num_resamples = 100)
    first_order_indices[idx, :] = Si["S1"]
    total_effect_indices[idx, :] = Si["ST"]
```

```python
fig, axes = plt.subplots(nrows =2, ncols = 4, figsize=(14, 12), dpi= 600)

axes[0,0].plot(saltelli_param_set, first_order_indices[:,0], label = "Aspect Ratio")

axes[0,1].plot(saltelli_param_set, first_order_indices[:,1], label = "Wind Speed")

axes[0,2].plot(saltelli_param_set, first_order_indices[:,2], label = "Wind Angle")

axes[0,3].plot(saltelli_param_set, first_order_indices[:,3], label = "Design Speed")

axes[0,0].set_ylabel("First Order Sobol Indices")
axes[1,0].set_ylabel("Total Effect Sobol Indices")

axes[1,0].plot(saltelli_param_set, total_effect_indices[:,0], label = "Aspect Ratio")

axes[1,1].plot(saltelli_param_set, total_effect_indices[:,1], label = "Wind Speed")

axes[1,2].plot(saltelli_param_set, total_effect_indices[:,2], label = "Wind Angle")

axes[1,3].plot(saltelli_param_set, total_effect_indices[:,3], label = "Design Speed")

for row in axes:
    for col in row:
        col.set_xlabel("Saltelli Parameter")
        col.legend()
```

### Fourier Amplitude Sensitivity Test

Lets try the same analysis using the Fourier Amplitude Sensitivity Test (FAST)

```python
dims_of_parameter_space = 4
problem = {'num_vars': dims_of_parameter_space,
          'names' : ["Aspect Ratio", "Wind Speed", "Wind Angle", "Design Speed"],
          'bounds': [[0, 2],
                    [120, 300],      # Note: These bounds are defined in FEM_Model_gable_roof(...)
                    [0, np.pi/2],
                    [100, 160]]
          }


fast_param_set = range(100, 5000, 50)

start_num = len(fast_param_set)

first_order_indices = np.zeros([start_num, 4])
second_order_indices = np.zeros([start_num, 4])
total_effect_indices = np.zeros([start_num, 4])


for idx, param in enumerate(fast_param_set):
    fast_parameter_number = param           
    param_values = fast_sampler.sample(problem, fast_parameter_number)
    damage_set = FEM_Model_gable_roof(param_values)  
    Si_fast = fast.analyze(problem, damage_set)
    first_order_indices[idx, :] = Si_fast["S1"]
    total_effect_indices[idx, :] = Si_fast["ST"]
```

```python
fig, axes = plt.subplots(nrows =2, ncols = 4, figsize=(14, 12), dpi= 600)

axes[0,0].plot(fast_param_set, first_order_indices[:,0], label = "Aspect Ratio")

axes[0,1].plot(fast_param_set, first_order_indices[:,1], label = "Wind Speed")

axes[0,2].plot(fast_param_set, first_order_indices[:,2], label = "Wind Angle")

axes[0,3].plot(fast_param_set, first_order_indices[:,3], label = "Design Speed")

axes[0,0].set_ylabel("First Order Sobol Indices")
axes[1,0].set_ylabel("Total Effect Sobol Indices")

axes[1,0].plot(fast_param_set, total_effect_indices[:,0], label = "Aspect Ratio")

axes[1,1].plot(fast_param_set, total_effect_indices[:,1], label = "Wind Speed")

axes[1,2].plot(fast_param_set, total_effect_indices[:,2], label = "Wind Angle")

axes[1,3].plot(fast_param_set, total_effect_indices[:,3], label = "Design Speed")

for row in axes:
    for col in row:
        col.set_xlabel("Saltelli Parameter")
        col.legend()
```

```python
Si_fast_df = Si_fast.to_df()

fig, axarr = plt.subplots(1, 3, figsize = [16, 8], dpi = 600)

plt.sca(axarr[0])
Si_fast_df["ST"].plot.bar(ax = axarr[0])
plt.title("FAST: Sub Domain, Small Sample Size")

plt.sca(axarr[1])
Si_new_df[0].plot.bar(ax = axarr[1])

plt.title("Sobol: Sub Domain, Small Sample Size")

plt.sca(axarr[1])
Si_df[0].plot.bar(ax = axarr[2])

plt.title("Sobol: Entire Domain, Large Sample Size")
```

```python
parameter_number_set = range(1000, 20000, 1000)        

n_samples_fast = np.zeros(len(parameter_number_set))

n_samples_saltelli = np.zeros_like(n_samples_fast)

for idx, parameter_number in enumerate(parameter_number_set):
    param_values_fast = fast_sampler.sample(problem, parameter_number)

    param_values_saltelli = saltelli.sample(problem, parameter_number)
    
    n_samples_fast[idx] = param_values_fast.shape[0]
    n_samples_saltelli[idx] = param_values_saltelli.shape[0]
```

```python
fig = plt.figure(figsize = [10, 6], dpi = 600)
plt.plot(parameter_number_set, n_samples_saltelli, label = "Saltelli")
plt.plot(parameter_number_set, n_samples_fast, label = "FAST")
plt.xlabel("Sampling Parameter")
plt.ylabel("Number of Samples Computed")
plt.legend()
plt.show()
```
