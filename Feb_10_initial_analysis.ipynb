```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import glob
```

```python
file_names = [name for name in glob.glob("*.csv")]

file_names
```

```python
df_list = []

for name in file_names:
    try:
        df_list.append(pd.read_csv(name))
    except:
        print("Error reading in File: %s"%name)
    
```

```python
#from the documentation for Daily Data, the column names are as follows
column_names_hourly = ["STATION_ID", "DATE", "HOUR", "JULIAN_DATE", "QC_ETo", 'REF_ETo', 
          "QC_PRECIP", "PRECIP", "QC_SOLAR_RAD", "SOLAR_RAD", "QC_VAPOR_PRESSURE",
          "VAPOR_PRESSURE", "QC_AIR TEMP", "AIR_TEMP", "QC_REL_HUMIDITY", "REL_HUMIDITY",
          "QC_DEW_POINT", "DEW_POINT", "QC_WIND_SPEED", "WIND_SPEED", "QC_WIND_DIRECTION", "WIND_DIRECTION",
          "QC_SOIL_TEMP", "SOIL_TEMP"]


column_names_daily = ["STATION_ID", 
                      "DATE", 
                      "JULIAN_DATE", 
                      "QC_SOL_RAD_AVG",
                      "SOL_RAD_AVG",
                      "QC_SOIL_TEMP_AVG",
                      "SOIL_TEMP_AVG",
                      "QC_AIR_TEMP_MAX",
                      "AIR_TEMP_MAX",
                      "QC_AIR_TEMP_MIN",
                      "AIR_TEMP_MIN",
                      "QC_AIR_TEMP_AVG",
                      "AIR_TEMP_AVG",
                      "QC_VAP_PRESS_AVG",
                      "VAP_PRESS_AVG",
                      "QC_WIND_SPEED_AVG",
                      "WIND_SPEED_AVG",
                      "QC_PRECIP",
                      "PRECIP",
                      "QC_REL_HUMID_MAX",
                      "REL_HUMID_MAX",
                      "QC_REL_HUMID_MIN",
                      "REL_HUMID_MIN",
                      "QC_REF_ETo",
                      "REF_ETo",
                      "QC_REL_HUMID_AVG",
                      "REL_HUMID_AVG",
                      "QC_DEW_POINT",
                      "DEW_POINT",
                      "QC_WIND_RUN",
                      "WIND_RUN"
                     ]

for idx, df in enumerate(df_list):
    df_list[idx].columns = column_names_daily
```

```python
dataset3 = dataset2.replace("--",np.nan, regex =True)
```

```python
for col in dataset3.columns:
    if (col[0:2] == "QC"):
        dataset3[col] = dataset3[col].astype('category')
    elif (col == "DATE") or (col == "JULIAN_DATE"):
        next
    else:
        dataset3[col] = pd.to_numeric(dataset3[col])
        
```

```python
dataset3.info(memory_usage="deep", verbose=True)
```

```python
station_list = pd.read_csv("CIMIS_Stations_List_January20.csv")
station_list.describe()
```

```python
station_list.head()
```

```python
station_num_to_lat = {}
station_num_to_long = {}
for num in station_list["Station Number"]:
    if ~np.isnan(num):
        mask = station_list["Station Number"] == num
        new_df = station_list[mask]
        station_num_to_lat["%d"%num] = new_df[mask]["Latitude"].values[0]
        station_num_to_long["%d"%num] = new_df[mask]["Longitude"].values[0]
```

```python
station_num_to_lat
```

```python
station_num_to_long
```

```python
dataset4 = dataset3.copy(deep = True)
dataset4["LATITUDE"] = dataset3["STATION_ID"].apply(lambda x: station_num_to_lat["%d"%x])
```

```python
dataset4["LONGITUDE"] = dataset3["STATION_ID"].apply(lambda x: station_num_to_long["%d"%x])
```

```python
dataset4.to_csv("Merged_Datasets.csv")
```
